```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import style
style.use('ggplot')
from sklearn.cluster import KMeans
```

```python
# plotting and visualising our data before feeding it into the machine learning algorithm
x = [1, 5, 1.5, 8, 1, 9]
y = [2, 8, 1.8, 8, 0.6, 11]

plt.scatter(x,y)
plt.show()
```

```python
# convert our data into a numpy array
data = []
for i in range(6):
    data.append([x[i], y[i]])
print(data)
X = np.array(data)
print(X)
```

```python
# initialize k-means algorithm with the required parameter and we use .fit() to fit the data.
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
# getting th evalues of centroids and lables based on the fitment
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
print("centroids :: \n{}\n".format(centroids))
print("labels    :: \n{}".format(labels))
```

```python
# plot and visualize the output.
colors = ['r.', 'g.', 'c.', 'y.']
for i in range (len(x)):
    print("coordinate: %10s \tlabel: %d" % (X[i], labels[i]))
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)

plt.scatter(
    centroids[:, 0], 
    centroids[:, 1], 
    marker = "x", 
    s = 150,
    linewidths = 5,
    zorder=10)
plt.show()
```
